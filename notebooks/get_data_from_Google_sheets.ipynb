```python
#AUTHOMATIC METHOD
# !pip install gspread google-auth
import gspread
from google.oauth2.service_account import Credentials

#For this method it is needed to create a service account in the Google Cloud Console, and to download the credentials.json file. 
#It is also needed to share the google sheet with the email of the service account, which can be found in the credentials.json file.
# It is a tedious process, but it allows to have the data updated automatically.

scopes = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]

creds = Credentials.from_service_account_file(
    "credentials.json",
    scopes=scopes
)

client = gspread.authorize(creds)
sheet = client.open_by_url("https://docs.google.com/spreadsheets/d/1ERMeyCK1gKepeToE2TkwSuv_xYyQbggwaCYIIp3k-Y0/edit?gid=0#gid=0").sheet1

print(sheet.get_all_records())
```
